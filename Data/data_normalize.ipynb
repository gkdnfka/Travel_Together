```python
import pandas as pd

df = pd.read_csv("./survey_data_after.csv", encoding='euc-kr')
```

```python
df
```

```python
import numpy as np

gender_data = df.loc[:, '성별']
want_data = df.loc[:, '경험']
#exp_data = df.loc[:, '경험']
personal_data = np.zeros(shape = (gender_data.size, 7))

for i in range(gender_data.size):
    for j in range(1, df.loc[i].size - 1):
        personal_data[i][j-1] = df.iloc[i][j]
personal_data /= 5
```

```python
want_hot_vec = np.zeros(shape = (want_data.size, 7))
want_dic = {'휴양':0, '자연경관':1, '문화유산': 2, '액티비티':3, '쇼핑':4, '공연':5, '전시물':6}

for i, data in zip(range(want_data.size), want_data):
    cat = data.split(', ')
    for j in range(len(cat)):
        want_hot_vec[i][want_dic[cat[j]]] = 1
```

```python
type(want_hot_vec[0])
```

```python
input_arr = []
for i in range(gender_data.size):
    input_arr.append(personal_data[i].copy())
    input_arr[i] = np.append(input_arr[i], 0 if gender_data[i] == '여성' else 1)
    
```

```python
new_df = pd.DataFrame(columns = ['input_data', 'want_hot_vec'])
for i in range(gender_data.size):
    new_df.loc[i] = [input_arr[i], want_hot_vec[i]]
```

```python
new_df.to_csv("./survey_norm.csv", index = False)
```
